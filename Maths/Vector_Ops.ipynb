```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
plt.quiver(0, 0, 4, 5)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0, -5, 4, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Addition of Two Vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])

sum = np.add(vector_1, vector_2)
print(sum)
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0, 0, 5, 1, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Subtraction of Two Vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])

diff = np.subtract(vector_1, vector_2)
print(diff)
```

Vector multiplication with scaler

```python
vector_1 = np.asarray([0, 0, 2, 3])
scalar = 3

product = np.multiply(vector_1, scalar)
print(product)
```

Dot Product of Two vectors

```python
a = np.array([2, 3])
b = np.array([3, 2])

dot_product = np.dot(a, b)
print(dot_product)
```

Cross Product

```python
c = np.array([2, 3, 11])
d = np.array([3, 2, 9])
cross_product = np.cross(c, d)
print(cross_product)
```

Projection of vector a on v

```python
a = np.array([2, 5])
v = np.array([8, -6])

projection = np.dot(a, v) / np.dot(v, v) * v
print(projection)
```
