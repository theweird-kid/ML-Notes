**Imbalanced Dataset:**
 dataset with an unequal class Distribution

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
credit_df = pd.read_csv('credit_data.csv')
```

```python
credit_df.head()
```

```python
credit_df.shape
```

```python
# distribution of the two classes
credit_df['Class'].value_counts()
```

This is highly Imabalanced Dataset
* 0 -> Legit Transactions
* 1 -> Fraudlent Transactions

```python
# Separating the legit and Fradulent Transactions
legit = credit_df[credit_df.Class == 0]
fraud = credit_df[credit_df.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

Undersampling to Handle Imbalanced Data

Building a sampled dataset containing similar Distribution of legit and Fraudlent Transactions

Number of Fraudlent Tx -> 492
Number of Legit Tx -> 284315

```python
legit_sample = legit.sample(n=492)
```

Concatenate the 2 DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
