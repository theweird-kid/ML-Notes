Label Encoding:

*   Converting the labels into numeric form



```python
# Importing the Deps
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# Loading the dataset
cancer_df = pd.read_csv('/content/data.csv')
```

```python
cancer_df.head()
```

```python
# finding the count of different labels
cancer_df['diagnosis'].value_counts()
```

```python
# Load the Label Encoder Function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_df.diagnosis)
```

```python
print(labels)
```

```python
# appending the labels to the DataFrame
cancer_df['target'] = labels
```

```python
cancer_df.head()
```



*   0 -> Benign
*   1 -> Malignant



```python
cancer_df['target'].value_counts()
```
