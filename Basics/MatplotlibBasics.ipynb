```python
# Import Matplotlib
import matplotlib.pyplot as plt
```

```python
# import numpy
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(y)
```

Plotting the Data

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding tile, x-axis, and y-axis labels
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title("Sine wave")
plt.show()
```

```python
# plotting parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('x^2')
plt.title("Parabola")
```

```python
# Plot with different symbols
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
```

```python
# Plotting multiple functions in same graph
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
# BAR PLOT
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 40, 50, 140, 59]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('Number of People')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 40, 50, 140, 59]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
