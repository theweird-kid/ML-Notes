```python
# Importing the Pandas library
import pandas as pd
```

```python
# importing the boston house price data
from sklearn.datasets import fetch_california_housing

california_hosing = fetch_california_housing()
#type(california_hosing)
print(california_hosing)
```

```python
# Creating a pandas dataframe
california_df = pd.DataFrame(california_hosing.data, columns = california_hosing.feature_names)
```
